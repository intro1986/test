```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Векторы
norm = np.random.normal(loc=0, scale=1, size=1000)
binom = np.random.binomial(n=10, p=0.5, size=1000)
unif = np.random.uniform(low=0, high=1, size=1000)
pois = np.random.poisson(lam=1, size=1000)
exp = np.random.exponential(scale=1, size=1000)
```

```python
# DataFrame
df = pd.DataFrame({'norm': norm, 'binom': binom, 'unif': unif, 'pois': pois, 'exp': exp})
df.head()
```

```python
# Рисуем гистограммы
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
fig.suptitle('Гистограммы распределений')
ax[0, 0].hist(norm, bins=30, alpha=0.5, label='norm', color='blue')
ax[0, 0].legend()
ax[0, 1].hist(binom, bins=20, alpha=0.5, label='binom', color='green')
ax[0, 1].legend()
ax[1, 0].hist(unif, bins=30, alpha=0.5, label='unif', color='red')
ax[1, 0].legend()
ax[1, 1].hist(pois, bins=20, alpha=0.5, label='pois', color='orange')
ax[1, 1].legend()
```
