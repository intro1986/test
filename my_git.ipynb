```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Векторы
norm = np.random.normal(loc=0, scale=1, size=100)
binom = np.random.binomial(n=10, p=0.5, size=100)
unif = np.random.uniform(low=0, high=1, size=100)
pois = np.random.poisson(lam=1, size=100)
exp = np.random.exponential(scale=1, size=100)
```

```python
# DataFrame
df = pd.DataFrame({'norm': norm, 'binom': binom, 'unif': unif, 'pois': pois, 'exp': exp})
df.head()
```
